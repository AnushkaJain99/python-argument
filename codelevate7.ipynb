```python
#types of argumaent in python
# 1. Positional argument

def calculate_volume(l, w, h):
    volume = l * w * h
    return volume
volume = calculate_volume(10, 5, 2)
print("The volume of the rectangular prism is",volume,"cubic units.")
```

```python
# 2.Default Argument

def calculate_volume(l, w=5, h=2):
    volume = l * w * h
    return volume
volume = calculate_volume(10)
print("The volume of the rectangular prism is",volume,"cubic units.")
```

```python
def sample(name,age=18):

  print("Hey!",name,"you are",age,"years old")
sample("Anushka")
```

```python
# 3.Keyword argument

def sample(name,age):

  print("Hey!",name,"you are",age,"years old")
sample(age=19,name="Aarushi")
sample(age=18,name="Anushka")
```

```python
# fibonacci series using recursion

def fibonacci(n):
    if n <= 0:
        return "please enter positive integer"
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

n = int(input("enter number"))
for i in range(1, n + 1):
    print(fibonacci(i),end=" ")

```
